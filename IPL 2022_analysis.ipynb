# Toss decision

```python
dec=sns.countplot(x ='toss_decision', data = df,width=0.3)
for bars in dec.containers:
    dec.bar_label(bars)
```

**As we can see most of the teams who have won the toss had opted for bowling**

# Top scorer for the team 

```python
dec=sns.countplot(x ='top_scorer',width=0.5, data = df)
sns.set(rc={'figure.figsize':(15,5)})
plt.xticks(rotation='vertical')   
plt.xlabel("Highest_scorer")
```

**jos buttler was the highest Run scorer for his team**

# Man of the match

```python
dec=sns.countplot(x ='player_of_the_match', data = df)
plt.xticks(rotation= 'vertical')
sns.set(rc={'figure.figsize':(15,8)})
plt.title('Man of the match',fontsize=15)
for bars in dec.containers:
    dec.bar_label(bars)
```

**kuldeep Yadav got most number of man of the match award**

# Highest individual score

```python
x=np.array(df['top_scorer'])
y=np.array(df['highscore'])
plt.figure(figsize=(15,5))
plt.xticks(rotation= 90)
plt.xlabel('high score')
hs=plt.bar(x,y)
plt.bar_label(hs,labels=y)

plt.show()
```

**Quinton de kock scored the highest individual score of the ipl scoring 140 runs and the next best was jos buttler with 116**

# Average Runs scored on the each venue

```python
ven=['Wankhede Stadium, Mumbai','Brabourne Stadium, Mumbai','Dr DY Patil Sports Academy, Mumbai','Maharashtra Cricket Association Stadium,Pune','Eden Gardens, Kolkata','Narendra Modi Stadium, Ahmedabad']
```

```python

avg1=df.groupby(['venue'],as_index=False)['first_ings_score'].mean().sort_values('first_ings_score',ascending=True)
avg2=df.groupby(['venue'],as_index=False)['second_ings_score'].mean().sort_values('second_ings_score',ascending=True)
p=np.arange(len(avg1))
width=0.2
p1=[i+width for i in p]
plt.xlabel('venue')
plt.ylabel('scores')
plt.title('average scores')


#labels,title etc
plt.bar(data=avg1,x=p,height='first_ings_score',color='r',width=0.2,label='average_first_ings_score')
plt.bar(data=avg2,x=p1,height='second_ings_score',width=0.2,label='average_first_ings_score')
plt.legend()
plt.xticks(p1,ven,rotation=90)
plt.figure(figsize=(15,5))
plt.show()
```

```python
result=df.groupby(['match_winner'],as_index=False)['toss_decision'].count().sort_values('toss_decision',ascending=True)
```

```python
plt.figure(figsize=(10,5))
plt.xlabel('toss_won_by_Teams')
plt.ylabel('match_won')
plt.title('toss and match won',fontsize=15)
plt.scatter(x='match_winner',y='toss_decision',data=result)
```

# Best_bowling figure

```python
best_bowler=df.groupby(['bowler_name'],as_index=False)["best_bowling_figure"].count().sort_values("best_bowling_figure",ascending=True)
```

```python
plt.figure(figsize=(15,8))
plt.bar(x='bowler_name',height="best_bowling_figure",data=best_bowler)
plt.xlabel('bowler_name')
plt.ylabel("highest wickettaker in the match")
plt.title('best_bowler',fontsize=15)
plt.xticks(rotation=90)
plt.show()
```
